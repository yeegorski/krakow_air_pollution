```python
# import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline 
```

```python
# read the preprocessed data
df = pd.read_csv('Weather_Smog_Data_Preprocessed_1.csv',index_col=0)
```

```python
df.head(10)
```

```python
# distribustion of PM2
df['PM2'].plot.hist(bins=30);
```

```python
sns.boxplot(x=df['PM2']);
```

```python
df[['PM2']].describe()
```

```python
# convert dates to datetime format
df['time'] = pd.to_datetime(df['time'],format = '%Y-%m-%d %H:%M:%S')
```

```python
# tempature vs date plus smog levels
fig, ax = plt.subplots(figsize=(15,5))
ax.set_xlim(xmin=df['time'].iloc[0], xmax=df['time'].iloc[-1])
sns.scatterplot(ax=ax,x = df['time'], y = df['PM2'], alpha=0.2, palette = 'rocket_r')
sns.lineplot(ax=ax, x= df['time'], y = np.full((df.shape[0],), 12), color='g')
sns.lineplot(ax=ax, x= df['time'], y = np.full((df.shape[0],), 35.4), color='y')
sns.lineplot(ax=ax, x= df['time'], y = np.full((df.shape[0],), 55.4), color='orange')
sns.lineplot(ax=ax, x= df['time'], y = np.full((df.shape[0],), 150.4), color='r')
sns.lineplot(ax=ax, x= df['time'], y = np.full((df.shape[0],), 250.4), color='purple');
# some outliers (exceeding PM2 level of 150) might need to be excluded for the linear model
```

```python
# tempature vs date plus smog levels
fig, ax = plt.subplots(figsize=(15,5))
ax.set_xlim(xmin=df['time'].iloc[0], xmax=df['time'].iloc[-1])
sns.scatterplot(ax=ax,x = df['time'], y = df['temperature'],hue = df['PM2'], alpha=0.2, palette = 'rocket_r');
```

```python
# dew point vs date plus smog levels
fig, ax = plt.subplots(figsize=(15,5))
ax.set_xlim(xmin=df['time'].iloc[0], xmax=df['time'].iloc[-1])
sns.scatterplot(ax=ax,x = df['time'], y = df['dewPoint'],hue = df['PM2'], alpha=0.2, palette = 'rocket_r');
```

```python
# humidity vs date plus smog levels
fig, ax = plt.subplots(figsize=(15,5))
ax.set_xlim(xmin=df['time'].iloc[0], xmax=df['time'].iloc[-1])
sns.scatterplot(ax=ax,x = df['time'], y = df['humidity'],hue = df['PM2'], alpha=0.2, palette = 'rocket_r');
```

```python
# pressure vs date plus smog levels
fig, ax = plt.subplots(figsize=(15,5))
ax.set_xlim(xmin=df['time'].iloc[0], xmax=df['time'].iloc[-1])
sns.scatterplot(ax=ax,x = df['time'], y = df['pressure'],hue = df['PM2'], alpha=0.2, palette = 'rocket_r');
```

```python
# wind speed vs date plus smog levels
fig, ax = plt.subplots(figsize=(15,5))
ax.set_xlim(xmin=df['time'].iloc[0], xmax=df['time'].iloc[-1])
sns.scatterplot(ax=ax,x = df['time'], y = df['windSpeed'],hue = df['PM2'], alpha=0.2, palette = 'rocket_r');
```

```python
# windGust vs date plus smog levels
fig, ax = plt.subplots(figsize=(15,5))
ax.set_xlim(xmin=df['time'].iloc[0], xmax=df['time'].iloc[-1])
sns.scatterplot(ax=ax,x = df['time'], y = df['windGust'],hue = df['PM2'], alpha=0.2, palette = 'rocket_r');
```

```python
# visibility vs date plus smog levels
fig, ax = plt.subplots(figsize=(15,5))
ax.set_xlim(xmin=df['time'].iloc[0], xmax=df['time'].iloc[-1])
sns.scatterplot(ax=ax,x = df['time'], y = df['visibility'],hue = df['PM2'], alpha=0.2, palette = 'rocket_r');
```

```python
# ozone vs date plus smog levels
fig, ax = plt.subplots(figsize=(15,5))
ax.set_xlim(xmin=df['time'].iloc[0], xmax=df['time'].iloc[-1])
sns.scatterplot(ax=ax,x = df['time'], y = df['ozone'],hue = df['PM2'], alpha=0.2, palette = 'rocket_r');
```

```python
# check relationship betwwen each pair of features
sns.pairplot(df)
```

```python
# create the heatmap of feature correlations
fig, ax = plt.subplots(figsize=(12,10))
sns.heatmap(df.corr(), annot = True, ax = ax)
```
